```python
from numpy import fft, arange, cos, pi, sin
from scipy import fftpack
import scipy
from matplotlib import pyplot as plt
import numpy as np
```

```python
t = np.linspace(0.1, 3*pi, 100)
signal1 = cos(t)
signal2 = sin(t)

n1 = signal1.size # window length signal 1
n2 = signal2.size # window length signal 2
d = t[1]-t[0] # sample spacing

fft1 = abs(scipy.fft(signal1))
fft2 = abs(scipy.fft(signal2))
freq1 = scipy.fftpack.fftfreq(n1, d)
freq2 = scipy.fftpack.fftfreq(n2, d)

fig = plt.figure(figsize=(12,6))
plt.subplot(211)
plt.plot(t, signal1)
plt.plot(t, signal2)
plt.legend(['cos', 'sin'], loc=4)

plt.subplot(212)
plt.plot(freq1,20*scipy.log10(fft1),'x')
plt.plot(freq2,20*scipy.log10(fft2),'x')
plt.legend(['fft-cos', 'fft-sin'])
#plt.xlim([-1,1])
plt.show()
```

```python
samplingFrequency = 32.
frequency = 4                           #frequency of the sinus wave
t = arange(0,20,1/samplingFrequency)    #time intervals with period 1/sampling frequency
y = cos(2*pi*frequency*t)

Y = fft.fft(y)                          #standard fft on the whole signal
frequencyAxis = fft.fftfreq(len(Y),1/samplingFrequency )          #adjusting the x axis 

#PLOTTING
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)
ax1.specgram(y, Fs = samplingFrequency)
ax2.stem(frequencyAxis,Y,linefmt='r--', markerfmt='ro')
```

```python
import scipy
import scipy.fftpack
import pylab
from scipy import pi
t = scipy.linspace(0,120,4000)
acc = lambda t: 5*scipy.sin(2*pi*5.0*t) + 5*scipy.sin(2*pi*10.0*t) + 2*scipy.random.random(len(t))

signal = acc(t)

FFT = abs(scipy.fft(signal))
freqs = scipy.fftpack.fftfreq(signal.size, t[1]-t[0])

fig = plt.figure(figsize=(12,6))
pylab.subplot(211)
pylab.plot(t, signal)
pylab.subplot(212)
pylab.plot(freqs,20*scipy.log10(FFT),'-')
pylab.show()
```
