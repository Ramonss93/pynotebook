```python
for m in range(2000,2014):
    year = []
    for n in files:
        if int(n[-11:-7]) == m:
            year.append(n)
    print (m)
    #print (len(year))
    
#    if len(year)== 365:
    for i in range(8, 369,8):
        
        # prepare for next for loop
        # first set range
        
        start = i - 8
        end = i
        #print ('i', i)
        
        # then set empy base array plus counter            
        base_array = np.full_like(arr, nan)
        base_array = base_array[None]
        #base_array = np.empty_like(arr, dtype=float)
        #base_array = np.ma.masked_equal(base_array, -99999.)
        count = 0
        
        for j in range(start,end):
            
            # first try if raster is available
            # what shall we do? sumation or mean? now it's mean
            
            try:
                #print (year[j])
                array = gdal.Open(year[j]).ReadAsArray()
                array = array[None]
                base_array = np.vstack((base_array, array))
                #array = np.ma.masked_equal(array, -99999.)                
                #base_array += array                
            except IndexError:
                pass
            continue
                
        # mask nan values and calculate mean over axis 0
        base_array = np.ma.masked_values(base_array, nan)
        base_array = base_array.mean(axis=0)
        #base_array = base_array.sum(axis=0) # DO THIS for accumulation
        
        folder = r'D:\Data\GPCP_doy8DayMean//'
        filename = 'GPCP_'+str(m)+str(start+1).zfill(3)+'.tif'
        path = folder+filename
        print (path)
        #print(base_array.data)
        dikkelul.saveRaster(path, base_array.data, ds, 6, nan=nan)                

        base_array = None
        array = None
```
