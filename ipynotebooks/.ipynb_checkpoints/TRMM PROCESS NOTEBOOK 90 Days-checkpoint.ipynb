```python
from osgeo import gdal, ogr
import os
import numpy as np
import datetime
import subprocess as sp
from mattijn import dikkelul as dl
```

```python
def FilesFolder(inGSODFolder, format_end=''):
    """
    All files within one folder
    """
    st_wmo = [os.path.join(root, name)
               for root, dirs, files in os.walk(inGSODFolder)
                 for name in files                 
                 if name.endswith(format_end)]
    return st_wmo

def FilesFolder2(RootFolder, contains='', format_end=''):
    """
    All files within all subfolders
    """
    lists = [os.path.join(root, name)    
                 for root, dirs, files in os.walk(RootFolder)
                   for name in files
                   if contains in name
                     if name.endswith(format_end)]
    return lists
```

```python
def saveRaster(path, array, dsSource, datatype=3, formatraster="GTiff", nan=None): 
    """
    Datatypes:
    unknown = 0
    byte = 1
    unsigned int16 = 2
    signed int16 = 3
    unsigned int32 = 4
    signed int32 = 5
    float32 = 6
    float64 = 7
    complex int16 = 8
    complex int32 = 9
    complex float32 = 10
    complex float64 = 11
    float32 = 6, 
    signed int = 3
    
    Formatraster:
    GeoTIFF = GTiff
    Erdas = HFA (output = .img)
    OGC web map service = WMS
    png = PNG
    """
    # Set Driver
    format_ = formatraster #save as format
    driver = gdal.GetDriverByName( format_ )
    driver.Register()
    
    # Set Metadata for Raster output
    cols = 1440#dsSource.RasterXSize
    rows = 400#dsSource.RasterYSize
    bands = 1#dsSource.RasterCount
    datatype = datatype#band.DataType
    
    # Set Projection for Raster
    outDataset = driver.Create(path, cols, rows, bands, datatype)
    geoTransform = gs#dsSource.GetGeoTransform()
    outDataset.SetGeoTransform(geoTransform)
    proj = dsSource.GetProjection()
    outDataset.SetProjection(proj)
    
    # Write output to band 1 of new Raster and write NaN value
    outBand = outDataset.GetRasterBand(1)
    if nan != None:
        outBand.SetNoDataValue(nan)
    outBand.WriteArray(array) #save input array
    #outBand.WriteArray(dem)
    
    # Close and finalise newly created Raster
    #F_M01 = None
    outBand = None
    proj = None
    geoTransform = None
    outDataset = None
    driver = None
    datatype = None
    bands = None
    rows = None
    cols = None
    driver = None
    array = None
```

### RAW TO TIF FORMAT

```python
folder = r'H:\3B42.daily.v7'
outfolder = r'D:\Data\TRMM//'
listfiles = FilesFolder(folder, format_end='.bin')

ds = gdal.Open(r'D:\Data\GPCP_tif//GPCP_2000001.tif')
gs = (-179.875, 0.25, 0.0, 51.125, 0.0, -0.25)

for fil in listfiles:
    # open bin file and convert big endian to float representation
    fid = open(fil, 'rb')
    dim = np.fromfile(fid, dtype='>f')
    dim = dim.astype('float32')
    
    # resize to world file and set out path
    dim.resize([400,1440])
    outfile = fil[-27:-4]
    outpath = outfolder + outfile + '.tif'
    
    saveRaster(outpath,  dim, ds, datatype=6)
    fid = None
    dim = None
    outfile = None
    print (outpath)
    outpath = None   
```

### Quindecim-day sums (XV-Day Totals) for TRMM data

```python
path = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\RAW'
outFolder = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\60_Day_Period\60_DaySums//'

files = FilesFolder(path, '.tif')
daySum = 60

# Open base_array
ds = gdal.Open(files[0])
arr = ds.ReadAsArray()
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()
if nan == None:
    nan = -99999.0
base_array = np.full_like(arr, nan)
base_array.shape

# Set interation for day sum
for i in range(daySum, len(files)+1, 1):
    start = i - daySum
    end = i
    print ('i', i, 'start', start, 'end', end)

    base_array = np.full_like(arr, nan)
    base_array = base_array[None]

    # Sum every dayfiles of period set with daySum
    for j in range(start,end):

        # first try if raster is available
        # what shall we do? sumation or mean? now it's mean

        try:
            print (files[j])
            array = gdal.Open(files[j]).ReadAsArray()
            array = array[None]
            base_array = np.vstack((base_array, array))
            #array = np.ma.masked_equal(array, -99999.)                
            #base_array += array                
        except IndexError:
            pass
        continue

    # mask nan values and calculate mean over axis 0
    base_array = np.ma.masked_values(base_array, nan)
    base_array = base_array.sum(axis=0)
    #base_array = base_array.sum(axis=0) # DO THIS for accumulation

    filename = files[j][-19::]
    path = outFolder+filename
    print (path)
    #print(base_array.data)
    dl.saveRaster(path, base_array.data, ds, 6, nan=nan)

    base_array = None
    array = None    
```

### Standardized Precipitation Anomaly

```python
pathmean = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_AllMaxMinStdMean//15_DaySums_Anomaly_Mean.tif'
pathstd = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_AllMaxMinStdMean//15_DaySums_Anomaly_Std.tif'
pathano = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_Anomaly'
outFolder = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_StdAnomaly//'

fileano = dl.FilesFolder(pathano, '.tif')
#fileano = fileano[0:-46]
#print (fileano)

# open basefile
ds = gdal.Open(fileano[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

# open max and min files, mask where NaN
meanin = gdal.Open(pathmean).ReadAsArray()
meanin = np.ma.masked_values(meanin, nan)
stdin = gdal.Open(pathstd).ReadAsArray()
stdin = np.ma.masked_values(stdin, nan)

for i in fileano:
    # open anomaly and mask where NaN
    ano = gdal.Open(i).ReadAsArray()
    ano = np.ma.masked_values(ano, nan)
    
    # normalized anomaly index
    normano = (ano - meanin) / stdin
    
    print ('\nmax', normano.max(), 'min', normano.min(), 'std', normano.std(), 'mean', normano.mean())
    path = outFolder+'TRMM_CQ_'+i[-11::]
    print (path)
    
    dl.saveRaster(path, normano.data, ds, datatype=6, nan=nan)
```

### Maximum and minimum Normalised Precipitation Anomaly

```python
pathIn = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_StdAnomaly'
pathMax = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_AllMaxMinStdMean//StdAno_max.tif'
pathMin = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Chongqing\TRMM\15_Day_Period\15_DaySums_AllMaxMinStdMean//StdAno_min.tif'

files = dl.FilesFolder(pathIn)
ds = gdal.Open(files[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

# Open two basefiles, one for max one for min, mask arrays where NaN
max_ds = gdal.Open(files[0]).ReadAsArray()
max_ds = np.ma.masked_values(max_ds, nan)
min_ds = gdal.Open(files[0]).ReadAsArray()
min_ds = np.ma.masked_values(min_ds, nan)

for i in files:
    
    # Open file, mask where NaN. 
    ds_comp = gdal.Open(i).ReadAsArray()
    ds_comp = np.ma.masked_values(ds_comp, nan)
    
    # Compare two arrays and returns a new array containing the element-wise maxima
    max_ds = np.maximum(max_ds, ds_comp)
    min_ds = np.minimum(min_ds, ds_comp)
    
    print (max_ds.max(), min_ds.min())    
    
dl.saveRaster(pathMax, max_ds.data, ds, datatype=6, nan=nan)
dl.saveRaster(pathMin, min_ds.data, ds, datatype=6, nan=nan)
```

### Normalised Anomaly between -1 and 1

```python
pathmax = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\NDVI\NDVI_AllMaxMinStdMean//NormAno_max.tif'
pathmin = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\NDVI\NDVI_AllMaxMinStdMean//NormAno_min.tif'
pathano = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\NDVI\NDVI_NormAnomaly'
outFolder = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\NDVI\NDVI_StanNormAnomaly//'

fileano = dl.FilesFolder(pathano, '.tif')
#fileano = fileano[0:-46]
#print (fileano)

# open basefile
ds = gdal.Open(fileano[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

# open max and min files, mask where NaN
maxin = gdal.Open(pathmax).ReadAsArray()
maxin = np.ma.masked_values(maxin, nan)
minin = gdal.Open(pathmin).ReadAsArray()
minin = np.ma.masked_values(minin, nan)

for i in fileano:
    # open anomaly and mask where NaN
    ano = gdal.Open(i).ReadAsArray()
    ano = np.ma.masked_values(ano, nan)
    
    # normalized anomaly index
    normano = ano / (maxin-minin)
    
    print ('\nmax', normano.max(), 'min', normano.min(), 'std', normano.std(), 'mean', normano.mean())
    path = outFolder+i[-19::]
    print (path)
    
    dl.saveRaster(path, normano.data, ds, datatype=6, nan=nan)
```
