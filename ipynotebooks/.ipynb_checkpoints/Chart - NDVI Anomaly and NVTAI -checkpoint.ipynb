# NORM TYPE-I

```python
import os
import pandas as pd
from osgeo import gdal
import numpy as np
import datetime
import matplotlib.pyplot as plt

def listall(RootFolder):
    lists = [os.path.join(root, name)    
                 for root, dirs, files in os.walk(RootFolder)
                   for name in files
                   if 'GPCP' in name
                     if name.endswith('.tif')]
    return lists

# ______________________________________________________
# SET PIXEL ROW AND COLUMN

pixelrow = 45
pixelcolumn = 296

# ______________________________________________________
# ANNUAL ACCUMULATED MEAN

li = []
da = []
i = 0
path = r'D:\Data\GPCP_30acc_anomalyNorm'# removed 'NormTypeII'
#path = r'D:\Data\GPCP_30acc_AnnualAcc'
test = listall(path)
test = test[0:-46]
print (len(test))
print (test[0])
for t in test:    
    raster = gdal.Open(t, gdal.GA_ReadOnly)
    nameRas = raster.GetDescription()[-11:-4]
    date = datetime.datetime(int(nameRas[0:4]), 1, 1) + datetime.timedelta(int(nameRas[4:9]) - 1) # replaced "int(nameRas[0:4])" with "2010" first year then add doy
    # print date
    da.append(date)

    band = raster.GetRasterBand(1)
    array = band.ReadAsArray()    
    
    # print array[879,4142] # Grassland pixel
    li.append(array[pixelrow, pixelcolumn])
    
    array = None
    band = None
    raster = None
    #print (str(i))
    i = i + 1
    
seriesNDVI = pd.Series(data = li, index = da)
GPCPnormTypeI = seriesNDVI

maxy = seriesNDVI.max()

print ('max', seriesNDVI.max(), GPCPnormTypeI.plot() )
```

# VISUALIZATION OF TYPE-I AND TYPE-II

```python
import matplotlib.pyplot as plt
import numpy as np


#______________________________
# TYPE-I

t = GPCPnormTypeI.index
signal = GPCPnormTypeI

FFT = abs(scipy.fft(signal))
timestep = t[1] - t[0]
freqs = scipy.fftpack.fftfreq(signal.size, timestep.days)

fig = plt.figure(figsize=(14,10))

ax1 = fig.add_subplot(321)
ax1.bar(t,signal)
#ax1.fill_between(t, signal, where=signal>=0, facecolor='lightblue', interpolate=True)#, alpha=0.5)
#ax1.fill_between(t, signal, where=signal<=0, facecolor='orange', interpolate=True)#, alpha=0.5)
ax1.set_title('Standardized Anomaly')
ax1.grid(True)
ax1.axhline(0, color='grey', lw=2)

ax3 = fig.add_subplot(323)
ax3.acorr(signal, usevlines=True, normed=True, maxlags=50, lw=2)#, color='b')
ax3.set_title('Autocorrelation of the Standardized Anomaly')
ax3.grid(True)
ax3.axhline(0, color='grey', lw=2)

ax5 = fig.add_subplot(325)
ax5.plot(freqs,20*scipy.log10(FFT),"+", color='black')
ax5.set_title('Frequency representation of the Standardized Anomaly')
ax5.grid(True)
#ax3.axhline(-10, color='grey', lw=2)

#______________________________
# TYPE-II

t = GPCPnormTypeII.index
signal = GPCPnormTypeII

FFT = abs(scipy.fft(signal))
timestep = t[1] - t[0]
freqs = scipy.fftpack.fftfreq(signal.size, timestep.days)

#fig = plt.figure(figsize=(12,10))

ax2 = fig.add_subplot(322)
ax2.bar(t,signal)
#ax1.fill_between(t, signal, where=signal>=0, facecolor='lightblue', interpolate=True)#, alpha=0.5)
#ax1.fill_between(t, signal, where=signal<=0, facecolor='orange', interpolate=True)#, alpha=0.5)
ax2.set_title('Standardized Anomaly')
ax2.grid(True)
ax2.axhline(0, color='grey', lw=2)

ax4 = fig.add_subplot(324)
ax4.acorr(signal, usevlines=True, normed=True, maxlags=50, lw=2)#, color='b')
ax4.set_title('Autocorrelation of the Standardized Anomaly')
ax4.grid(True)
ax4.axhline(0, color='grey', lw=2)

ax6 = fig.add_subplot(326)
ax6.plot(freqs,20*scipy.log10(FFT),"+", color='black')
ax6.set_title('Frequency representation of the Standardized Anomaly')
ax6.grid(True)
#ax3.axhline(-10, color='grey', lw=2)

plt.tight_layout()
print('Visualization TYPE-II')
plt.show()
```
