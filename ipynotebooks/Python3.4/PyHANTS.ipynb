```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
class InputError( Exception ): pass
```

```python
"""
ni    = nr. of images (total number of actual samples of the time series)
nf    = number of frequencies to be considered above the zero frequency
"""
ni = 36
nf = 3
```

```python
# create empty arrays to fill
mat = np.zeros(shape=(min(2*nf+1,ni),ni))
amp = np.zeros(shape=(nf+1))
phi = np.zeros(shape=(nf+1))
yr = np.zeros(shape=(ni))
print ('mat',mat.shape, 
       '\namp',amp.shape, # returned array of amplitudes, first element is the average of the curve
       '\nphi',phi.shape, # returned array of phases, first element is zero
       '\nyr', yr.shape) # array holding reconstructed time series
```

```python
"""
HiLo  = 2-character string indicating rejection of high or low outliers
        select from 'Hi', 'Lo' or None
dod   = degree of overdeterminedness (iteration stops if number of 
        points reaches the minimum required for curve fitting, plus 
        dod). This is a safety measure
nb    = length of the base period, measured in virtual samples 
        (days, dekads, months, etc.)
"""
HiLo = 'Hi'
dod = 1 #Jie uses 2 * nf -1 # 2*4-1=7 # 2*-1 number of frequencies
nb = 36
```

```python
if HiLo:
    sHiLo = 0
    if HiLo == 'Hi':
        sHiLo =-1
    elif HiLo == 'Lo':
        sHiLo = 1
        
    nr = min(2*nf+1,ni) # number of 2*+1 frequecies, or number of input images
    noutmax = ni-nr-dod # number of input images - number of 2*+1 frequencies - degree of overdeterminedness
    dg = 180./np.pi
    mat[0,:] = 1    
    ang = 2*np.pi*np.arange(nb)/nb
    cs = np.cos(ang)
    sn = np.sin(ang)
```

```python
"""
ts    = array of size ni of time sample indicators 
        (indicates virtual sample number relative to the base period); 
        numbers in array ts maybe greater than nb
        If no aux file is used (no time samples), we assume ts(i)= i, 
        where i=1, ..., ni
"""
ts = np.arange(ni)
```

```python
#mat = np.zeros(shape=(min(2*nf+1,ni),ni))
#mat[0,:] = 1 
#print (mat, '\n',mat.shape)
```

```python
i = np.arange(1,nf+1)
for j in np.arange(ni):    
    index = np.mod(i*ts[j],nb)
    mat[2 * i-1,j] =cs[index]
    mat[2 * i ,j] = sn[index]    
```

```python
"""
y     = array of input sample values (e.g. NDVI values)
low   = valid range minimum
high  = valid range maximum (values outside the valid range are rejeced
        right away)
"""

y= np.array([5.0,  2.0,  5.0, 10.0, 12.0, 18.0, 20.0, 23.0, 27.0, 
             30.0, 40.0, 60.0, 66.0, 70.0, 90.0,120.0,160.0,190.0,
             105.0,210.0,104.0,200.0, 90.0,170.0, 50.0,120.0, 80.0,
             60.0, 50.0, 40.0, 30.0, 28.0, 24.0, 20.0, 15.0, 10.0 ])
low = 0.
high = 255.
```

```python
p = np.ones(ni)
p[(low > y) | (y > high)] = 0
nout = sum(p==0) # number of inputs imagery that are marked as outliers

if nout>noutmax:
    raise InputError("Not enough data points. Nout: "+str(nout)+" noutmax: "+str(noutmax)+". Nout > noutmax")
```

```python
for i in range(mat.shape[0]):
    plt.plot(mat[i])
```

```python
"""
delta = small positive number (e.g. 0.1) to suppress high amplitudes
"""
delta = 0.1
```

```python
# multipy outliers with timeseries
za = mat.dot((p * y))
# multiply mat with the multiplication of multiply diagonal of p with transpose of mat
A = mat.dot(np.diag(p).dot(mat.T))
# add delta to supress high amplitudes
A = A + np.diag(np.ones(nr)) * delta
# but remove the delta from [0,0]
A[0,0] = A[0,0] - delta
# solve linear matrix equation
zr = np.linalg.solve(A,za)

# first timeseries
yr[:] = mat.T.dot(zr)
# calculate error
diffVec = sHiLo * (yr-y)
err = p*diffVec

# sort errors
rankVec = np.argsort(err, axis=-1)

# select maximum error by 
maxerr = diffVec.take(rankVec.take(ni-1)) # same as diffVec[rankVec1[ni-1]] but 10x quicker
```

```python
plt.plot(y)
plt.plot(yr)
plt.show()
```

```python
"""
fet   = fit error tolerance (points deviating more than fet from curve 
        fit are rejected)
"""
fet = 5.0
```

```python
ready = (maxerr <= fet) | (nout == noutmax)
print (ready)
```

```python
if (not ready):
    i = ni # i is number of input images
    j = rankVec.take(i-1)    
    print ('kk', i,j)
    while ((p[j] * diffVec[j] > maxerr * 0.5) and (nout < noutmax)):

        p[j]=0
        nout=nout + 1
        i=i - 1
        j=np.linalg.matrix_rank(i)
        print ('oo',i,j)
```

```python
ready = False
nloop = 0
nloopmax = ni
print (nloopmax)
a = np.arange(nloopmax)
it = np.nditer(a)
while ((not it.finished) & (not ready)):
    print ("%d" % (it[0]))
    
    # multipy outliers with timeseries
    za = mat.dot((p * y))
    # multiply mat with the multiplication of multiply diagonal of p with transpose of mat
    A = mat.dot(np.diag(p).dot(mat.T))
    # add delta to supress high amplitudes
    A = A + np.diag(np.ones(nr)) * delta
    # but remove the delta from [0,0]
    A[0,0] = A[0,0] - delta
    # solve linear matrix equation
    zr = np.linalg.solve(A,za)

    # first timeseries
    yr[:] = mat.T.dot(zr)
    # calculate error
    diffVec = sHiLo * (yr-y)
    err = p*diffVec

    # sort errors
    rankVec = np.argsort(err, axis=-1)

    # select maximum error by 
    maxerr = diffVec.take(rankVec.take(ni-1)) # same as diffVec[rankVec1[ni-1]] but 10x quicker   
    
    ready = (maxerr <= fet) | (nout == noutmax)
    print (ready)
    plt.plot(y)
    plt.plot(yr)
    plt.show()
    
    if (not ready):
        i = ni # i is number of input images
        j = rankVec.take(i-1)    
        print ('kk', i,j)
        while ((p.take(j) * diffVec.take(j) > maxerr * 0.5) and (nout < noutmax)):
            
            # set highest error to 0, increase number of outliers
            p[j]=0
            nout=nout + 1
            # decrease number of input images
            i=i - 1
            # return matrix rank
            j=np.linalg.matrix_rank(i)
            print ('oo',i,j)
    
    it.iternext()
```

```python
amp[0] = zr[0]
phi[0] = 0

i=np.arange(1,nr,2)
ifr=((i+1)/2).astype(int)
ra=zr[i]
rb = zr[i+1]

amp[ifr] = np.sqrt(ra*ra+rb*rb)
phase = np.arctan2(rb,ra)*dg
phase[phase<0]=phase[phase<0]+360
phi[ifr]=phase
```
