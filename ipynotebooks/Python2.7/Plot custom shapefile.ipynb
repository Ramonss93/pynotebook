```python
bound = r'D:\Data\ChinaShapefile//ChinaProvBoundaryLine.shp'
capit = r'D:\Data\ChinaShapefile//ChinaProvCapitalsPoint.shp'
lakes = r'D:\Data\ChinaShapefile//ChinaLakesPolygon.shp'
river = r'D:\Data\ChinaShapefile//ChinaRiversLine.shp'
```

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.io.shapereader import Reader

ax = plt.axes(projection=ccrs.PlateCarree())
#ax.set_extent([73.43693268011174, 135.06693268011173, 53.57324111988825, 18.203241119888254])
#ax.set_extent([111.91693268, 123.85693268, 49.43324112, 40.67324112]) # InnerMongolia
#ax.set_extent([104.65693268, 110.61693267999999, 33.26324112, 27.763241120000004]) # Chongqing
ax.set_extent([113.87693268, 118.86693267999999, 31.173241119999997, 27.703241119999998]) # Jiujang

shape_bound = cfeature.ShapelyFeature(Reader(bound).geometries(), ccrs.PlateCarree(), facecolor='none')
shape_capit = cfeature.ShapelyFeature(Reader(capit).geometries(), ccrs.PlateCarree(), facecolor='none')
shape_lakes = cfeature.ShapelyFeature(Reader(lakes).geometries(), ccrs.PlateCarree(), facecolor='none')
shape_river = cfeature.ShapelyFeature(Reader(river).geometries(), ccrs.PlateCarree(), facecolor='none')

#ax.add_feature(cfeature.LAND, facecolor='0.9')
#ax.add_feature(cfeature.BORDERS, linewidth = '0.3')
ax.add_feature(shape_bound, linewidth='0.5', alpha='0.8')
ax.add_feature(shape_capit, linewidth='0.5', alpha='0.8')
ax.add_feature(shape_lakes, linewidth='0.5', alpha='0.8', facecolor='#96DBF2', color='#0A93FC')#'#96DBF2')
ax.add_feature(shape_river, linewidth='0.5', alpha='0.8', color='#0A93FC')

ax.add_feature(cfeature.OCEAN, facecolor='aqua', alpha='0.1')


ax.add_feature(cfeature.COASTLINE)
#ax.add_feature(states_provinces, edgecolor='gray')
#ax.coastlines(resolution='50m', linewidth='0.5')

plt.show()
```
