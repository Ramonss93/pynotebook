```python
import os
import pandas as pd
from osgeo import gdal
import numpy as np
import datetime
import matplotlib.pyplot as plt

def listall(RootFolder):
    lists = [os.path.join(root, name)    
                 for root, dirs, files in os.walk(RootFolder)
                   for name in files
                   if 'GPCP' in name
                     if name.endswith('.tif')]
    return lists

# ______________________________________________________
# SET PIXEL ROW AND COLUMN

pixelrow = 45
pixelcolumn = 296

# ______________________________________________________
# ANNUAL ACCUMULATED MEAN

li = []
da = []
i = 0
path = r'D:\Data\GPCP_30acc_anomalyNorm'# removed 'NormTypeII'
#path = r'D:\Data\GPCP_30acc_AnnualAcc'
test = listall(path)
test = test[0:-46]
print (len(test))
print (test[0])
for t in test:    
    raster = gdal.Open(t, gdal.GA_ReadOnly)
    nameRas = raster.GetDescription()[-11:-4]
    date = datetime.datetime(int(nameRas[0:4]), 1, 1) + datetime.timedelta(int(nameRas[4:9]) - 1) # replaced "int(nameRas[0:4])" with "2010" first year then add doy
    # print date
    da.append(date)

    band = raster.GetRasterBand(1)
    array = band.ReadAsArray()    
    
    # print array[879,4142] # Grassland pixel
    li.append(array[pixelrow, pixelcolumn])
    
    array = None
    band = None
    raster = None
    #print (str(i))
    i = i + 1
    
seriesNDVI = pd.Series(data = li, index = da)
GPCPnormTypeI = seriesNDVI

maxy = seriesNDVI.max()

print ('max', seriesNDVI.max(), GPCPnormTypeI.plot() )
```

```python
#!/usr/bin/env python
"""
Compute the coherence of two signals
"""
import numpy as np
import matplotlib.pyplot as plt

# make a little extra space between the subplots
plt.subplots_adjust(wspace=0.5)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2
r = np.exp(-t/0.05)

cnse1 = np.convolve(nse1, r, mode='same')*dt   # colored noise 1
cnse2 = np.convolve(nse2, r, mode='same')*dt   # colored noise 2

# two signals with a coherent part and a random part
s1 = 0.01*np.sin(2*np.pi*10*t) + cnse1
s2 = 0.01*np.sin(2*np.pi*10*t) + cnse2

plt.subplot(211)
plt.plot(NVAI_SPI.index, NVAI_SPI.NVAI, 'b-', NVAI_SPI, NVAI_SPI.SPI, 'g-')
plt.xlim(0,5)
plt.xlabel('time')
plt.ylabel('s1 and s2')
plt.grid(True)

plt.subplot(212)
cxy, f = plt.cohere(NVAI_SPI.NVAI, s2, 256, 1./dt)
plt.ylabel('coherence')
plt.tight_layout()
plt.show()
```
