```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.arange(100)
data = 2*np.sin(2*(np.pi/2.2)*t)+5*np.sin(2*(np.pi/1.02)*t)+2*np.sin(2*(np.pi/0.52)*t)

plt.plot(t,data)
```

```python
plt.psd(data, Fs=1)
plt.show()
```

```python
import numpy as np
sampling_length = 1 # measured every 15 minutes
Fs = 1.0/sampling_length
ls = range(len(data)) # data contains the function
freq = np.fft.fftfreq(len(data), d = sampling_length)
fft = np.fft.fft(data)
fft2 = np.fft.fft(data)
x = freq[:len(data)/2]
for i in range(len(x)):
    if x[i] < 0.4090000:
        # cut off all frequencies higher than 0.005
        fft[i] = 0.0
        fft[len(data)/2 + i] = 0.0
inverse = np.fft.ifft(fft)

for i in range(len(x)):
    if x[i] > 0.020000:
        # cut off all frequencies higher than 0.005
        fft2[i] = 0.0
        fft2[len(data)/2 + i] = 0.0
inverse2 = np.fft.ifft(fft2)
```

```python
plt.figure(figsize(13,3))
subplot(131)
plt.plot(t,data)
plt.grid()

subplot(132)
plt.psd(data, Fs=1)

subplot(133)
plt.plot(t,data)
plt.plot(t,inverse2-inverse)
plt.plot(t,inverse2)
plt.grid()
lg=plt.legend(['data','high','low'], loc=3, ncol=3)
lg.get_frame().set_lw(0.5)
lg.get_frame().set_ec('lightgray')
plt.tight_layout()
plt.show()
```
