```python
import datetime
import numpy as np
import pandas as pd
import matplotlib as mpl
import sys, os, re

# Colors via http://colorbrewer2.com/
WHITE, LIGHT_GRAY, GRAY, BLACK = [ "#FFFFFF", "#E5E5E5", "#777777", "#000000" ] 
COLORS = [ "#FF6600", "#D7191C", "#2C7BB6", "#FFFFBF", "#ABD9E9" ]
ORANGE, RED, BLUE, YELLOW, LIGHT_BLUE = COLORS
HIGH_ALPHA = 0.9
MEDIUM_ALPHA = 0.5
LOW_ALPHA = 0.1
rgba = mpl.colors.colorConverter.to_rgba

def set_styles(style_dict):
    """Set matplotlib styles from nested a nested dictionary"""
    for obj in style_dict: mpl.rc(obj, **style_dict[obj]) 

set_styles({
    "figure": { "figsize": [ 12, 8 ], "facecolor": WHITE },
    "savefig": { "dpi": 200, "bbox": "tight" },
    "patch": { "linewidth": 0.5, "facecolor": ORANGE, "edgecolor": WHITE, "antialiased": True },
    "font": { "size": 12 },
    "legend": { "fontsize": 10 },
    "axes": { 
        "facecolor": LIGHT_GRAY, 
        "edgecolor": WHITE, 
        "linewidth": 1, 
        "grid": False, 
        "titlesize": "large", 
        "labelsize": "large", 
        "labelcolor": GRAY,
        "axisbelow": True,
        "color_cycle": COLORS
    },
    "xtick": { "color": GRAY, "direction": "out" },
    "ytick": { "color": GRAY, "direction": "out" },
    "grid": { "color": WHITE, "linestyle": "-" }
})
```

```python
#____________
#style.use('ggplot')
ax1 = plt.subplot(111)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2009
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,625], color=BLUE, linewidth=2)
amean.plot(ax=ax1, ylim=[0,625], color=ORANGE, linewidth=2).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor=LIGHT_BLUE, interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='ORANGE', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)
#____________

outpath = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\TRMM2\5_Day_Period//Accumulated20009_'+str(pixelrow)+'C'+str(pixelcolumn)+'.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)
plt.show()
```
