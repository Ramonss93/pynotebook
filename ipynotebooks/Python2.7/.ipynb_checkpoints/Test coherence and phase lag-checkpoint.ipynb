```python
import numpy as np
import matplotlib.pyplot as plt
```

Create two random signals

```python
nfft = 128
dt = 1 # hour
t = np.arange(0,1774,dt)
rand1 = rand(len(t));
rand2 = rand(len(t));
s1 = 1*cos(2*pi*90/360+2*pi/21*t)+4*cos(+2*pi*0/360-2*pi/10*t)+1.5*(-1+2*rand1);
s2 = 1*cos(2*pi*60/360+2*pi/21*t)+4*cos(-2*pi*70/360-2*pi/10*t)+1.5*(-1+2*rand2);
```

```python
nfft = 256
dt = 0.01 # hour
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2
r = np.exp(-t/0.05)

cnse1 = np.convolve(nse1, r, mode='same')*dt   # colored noise 1
cnse2 = np.convolve(nse2, r, mode='same')*dt   # colored noise 2

# two signals with a coherent part and a random part
s1 = 0.01*np.sin(2*np.pi*10*t) + cnse1
s2 = 0.01*np.sin(2*np.pi*10*t) + cnse2
```

Plot the two signals

```python
plt.figure(figsize=(12,3))

subplot(121)
plt.plot(t,s1)
plt.plot(t,s2)
plt.ylabel('amplitude')
plt.xlabel('sample number')
plt.title('two series')
plt.grid(True)

subplot(122)
plt.plot(t,s1)
plt.plot(t,s2)
plt.ylabel('amplitude')
plt.xlabel('sample number')
plt.title('zoom of signals')
plt.xlim(0,100)
plt.grid(True)

plt.tight_layout()

outpath = r'C:\Users\lenovo\Pictures\cross-spectral-phaselag//series_zoom.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)
plt.show()
```

Power spectral density estimates for signal 1 and 2

```python
from pylab import *
plt.figure(figsize=(12,3))

#plot(t,s1)
subplot(121)
psd(s1, nfft, 1/dt)
plt.title('signal1')

subplot(122)
psd(s2, nfft, 1/dt)
plt.title('signal2')

plt.tight_layout()

outpath = r'C:\Users\lenovo\Pictures\cross-spectral-phaselag//psw_s1_s2.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)

show()
```

Cross (power) spectral density between signal 1 and 2

```python
plt.figure(figsize=(6,3))
csdxy, fcsd = plt.csd(s1, s2, nfft, 1./dt)
plt.ylabel('CSD (db)')
plt.title('cross spectral density between signal 1 and 2')
plt.tight_layout()

outpath = r'C:\Users\lenovo\Pictures\cross-spectral-phaselag//csd.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)

show()
```

Calculate phase using the cross (power) spectral density

```python
collect = []
for i in range(len(csdxy)):
    collect.append(np.math.atan2(-csdxy.imag[i],csdxy.real[i]))
col = np.array(collect)
phase = col * 180/np.pi
```

Plot of coherence between signal 1 and 2

```python

plt.figure(figsize=(12,3))
#plot(t,s1)
subplot(121)
cxy, fcoh = cohere(s1, s2, nfft, 1./dt)
plt.ylabel('Coherence')
plt.title('(magnitude) coherence between signal 1 and 2')

subplot(122)
plt.plot(fcoh,phase)
plt.ylabel('Phase (degrees)')
plt.xlabel('Frequency')
plt.ylim([-200,200])
plt.yticks([-180,-135,-90,-45,0,45,90,135,180])
plt.title('phase between signal 1 and 2')
plt.grid(True)

plt.tight_layout()
outpath = r'C:\Users\lenovo\Pictures\cross-spectral-phaselag//coherencephase_2plot.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)

show()
```

Plot of coherence and phase as functions of frequency with confidence level greater than 95%

```python
# calculate 95% confidence level
edof = (len(s1)/(nfft/2)) * cxy.mean() # equivalent degrees of freedom: (length(timeseries)/windowhalfwidth)*mean_coherence
gamma95 = 1.-(0.05)**(1./(edof-1.))
conf95 = np.where(cxy>gamma95)
print 'gamma95',gamma95, 'edof',edof

# Plot twin plot
fig, ax1 = plt.subplots()
# plot on ax1 the coherence
ax1.plot(fcoh, cxy, 'b-')
ax1.set_xlabel('Frequency (min^-1)')
ax1.set_ylim([0,1])
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Coherence', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')

# plot on ax2 the phase
ax2 = ax1.twinx()
ax2.plot(fcoh[conf95], phase[conf95], 'r.')
ax2.set_ylabel('Phase (degrees)', color='r')
ax2.set_ylim([-200,200])
ax2.set_yticklabels([-180,-135,-90,-45,0,45,90,135,180])

for tl in ax2.get_yticklabels():
    tl.set_color('r')

ax1.grid(True)
#ax2.grid(True)
fig.suptitle('Coherence and phase (>95%) between signal 1 and 2', fontsize='12')

outpath = r'C:\Users\lenovo\Pictures\cross-spectral-phaselag//coherencephase_singleplot.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)

plt.show()
```
