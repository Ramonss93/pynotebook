```python
import os
import pandas as pd
from osgeo import gdal
import numpy as np

def listall(RootFolder):
    lists = [os.path.join(root, name)    
                 for root, dirs, files in os.walk(RootFolder)
                   for name in files
                   if 'TRMM' in name
                     if name.endswith('.tif')]
    return lists

# ______________________________________________________
# SET PIXEL ROW AND COLUMN

#pixelrow = 37
#pixelcolumn = 179
pixelcolumn = 22 # extract from rasterXSize
pixelrow = 27 # extract from rasterYS

# ______________________________________________________
# ANNUAL ACCUMULATED MEAN

li = []
da = []
i = 0

path = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\TRMM2\5_Day_Period\5_DaySums_Accumulated'
test = listall(path)
print (len(test))
print (test[0])
for t in test:   
    print (t)
    raster = gdal.Open(t, gdal.GA_ReadOnly)
    nameRas = raster.GetDescription()[-11:-4]
    date = datetime.datetime(int(nameRas[0:4]), 1, 1) + datetime.timedelta(int(nameRas[4:9]) - 1) # replaced "int(nameRas[0:4])" with "2010" first year then add doy
    # print date
    da.append(date)

    band = raster.GetRasterBand(1)
    array = band.ReadAsArray()    
    
    # print array[879,4142] # Grassland pixel
    li.append(array[pixelrow, pixelcolumn])
    
    array = None
    band = None
    raster = None
    #print (str(i))
    i = i + 1
    
seriesNDVI = pd.Series(data = li, index = da)
y2003 = seriesNDVI[0:73]
y2004 = seriesNDVI[73:73+74]
y2005 = seriesNDVI[73+74:73+74+73]
y2006 = seriesNDVI[73+74+73:73+74+73+73]
y2007 = seriesNDVI[73+74+73+73:73+74+73+73+73]
y2008 = seriesNDVI[73+74+73+73+73:73+74+73+73+73+74]
y2009 = seriesNDVI[73+74+73+73+73+74:73+74+73+73+73+74+73]
y2010 = seriesNDVI[73+74+73+73+73+74+73:73+74+73+73+73+74+73+73]
y2011 = seriesNDVI[73+74+73+73+73+74+73+73:73+74+73+73+73+74+73+73+73]
y2012 = seriesNDVI[73+74+73+73+73+74+73+73+73:73+74+73+73+73+74+73+73+73+74]

#y2003 = seriesNDVI[0:25]
#y2004 = seriesNDVI[25:50]
#y2005 = seriesNDVI[50:75]
#y2006 = seriesNDVI[75:100]
#y2007 = seriesNDVI[100:125]
#y2008 = seriesNDVI[125:150]
#y2009 = seriesNDVI[150:175]
#y2010 = seriesNDVI[175:200]
#y2011 = seriesNDVI[200:225]
#y2012 = seriesNDVI[225:250]
#y2010 = seriesNDVI[460:506]
#y2011 = seriesNDVI[506:552]
#y2012 = seriesNDVI[552:598]
maxy = seriesNDVI.max()

print ('max', seriesNDVI.max(), seriesNDVI.plot() )

# ______________________________________________________
# ANNUAL ACCUMULATED lONG TERM MEAN

li = []
da = []
i = 0

path = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\TRMM2\5_Day_Period\5_DaySums_AnnualMean_Accumulated'
test = listall(path)
print (len(test))
print (test[0])
for t in test:    
    print (t)
    raster = gdal.Open(t, gdal.GA_ReadOnly)
    nameRas = raster.GetDescription()[-11:-4]
    date = datetime.datetime(2002, 1, 1) + datetime.timedelta(int(nameRas[4:9]) - 1) # replaced "int(nameRas[0:4])" with "2010" first year then add doy
    # print date
    da.append(date)

    band = raster.GetRasterBand(1)
    array = band.ReadAsArray()    
    
    # print array[879,4142] # Grassland pixel
    li.append(array[pixelrow,pixelcolumn])
    
    array = None
    band = None
    raster = None
    #print (str(i))
    i = i + 1

seriesGPCP03 = pd.Series(data = li, index = da)
#print (seriesGPCP03.head(), seriesGPCP03.plot() )

# ______________________________________________________
# FROM HERE PRINT FIGURES

fig = plt.figure(figsize=(26,15))


ax1 = plt.subplot(331)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2004
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(332)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2005
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(333)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2006
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(334)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2007
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(335)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2008
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(336)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2009
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________


ax1 = plt.subplot(337)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2010
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(338)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2011
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

ax1 = plt.subplot(339)
ax2 = ax1.twinx()

window = [1, 0, 0 , 0 -1] 
slope_ltm = np.convolve(li, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_ltm = np.hstack([np.ones(padlength), slope_ltm]) 

yy = y2012
window = [1, 0, 0 , 0 -1] 
slope_yy = np.convolve(yy, window, mode='valid') / float(len(window) - 1) 
padlength = len(window) -1 
slope_yy = np.hstack([np.ones(padlength), slope_yy]) 
slopePrc_YEAR = pd.Series(data = slope_yy, index = yy.index)
if len(li) == yy.size:
    amean = pd.Series(data = li, index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm
else:    
    amean = pd.Series(data = li[0:-1], index = yy.index)
    slopedt = slopePrc_YEAR - slope_ltm[0:-1]
axhline = np.zeros_like(slopedt)

yy.plot(ax=ax1, ylim=[0,maxy])
amean.plot(ax=ax1, ylim=[0,maxy]).set_title(str(yy.index.year[0]), fontsize=20)
slopedt.plot(ax=ax2, color='0.5', linestyle='-', secondary_y=True, ylim=([-50,50]), yticks=([-45,-30,-15,0,15,30,45]))
ax2.axhline(0.5, color='0.5', linestyle='-')

ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values>=amean.values, facecolor='lightblue', interpolate=True, alpha=0.5) 
ax1.fill_between(yy.index, yy.values, amean.values, where=yy.values<=amean.values, facecolor='orange', interpolate=True, alpha=0.5) 

ax1.set_ylabel('preciptation (mm)')
ax2.set_ylabel('slope (degrees)')

ax1.legend([str(yy.index.year[0]), 'longterm mean'],loc=2)
ax2.legend(['slope','test'],loc=4)

#____________

plt.tight_layout()
#plt.suptitle('GPCP Accumulation',y=-0.0, fontsize=28)
outpath = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\TRMM2\5_Day_Period//AccumulatedPrec_for_R'+str(pixelrow)+'C'+str(pixelcolumn)+'.png'
print ('saved to: ', outpath)
plt.savefig(outpath, dpi=200)
plt.show()
```
