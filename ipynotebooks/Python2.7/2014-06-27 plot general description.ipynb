```python
import numpy as np
import matplotlib.pyplot as plt

#http://matplotlib.sourceforge.net/examples/pylab_examples/spine_placement_demo.html

fig = plt.figure(figsize=(4, 5))

ax = fig.add_subplot(111)


x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.plot(x, y,linewidth="3", alpha=0.3)

ax.plot([0, np.pi/2], [1, 1], ls="--", color="green", linewidth="3",alpha=0.5)
ax.plot([np.pi/2, np.pi/2], [1, 0], ls="--", color="red", linewidth="3",alpha=0.5)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')


ax.set_xlim(-2*np.pi, 2*np.pi)

xticker = np.arange(-np.pi-np.pi/2, np.pi+np.pi, np.pi/2)
xlabels = [r"$\frac{-3\pi}{2}$", r"${-\pi}$",r"$\frac{-\pi}{2}$","",r"$\frac{pi}{2}$",r"${\pi}$",r"$\frac{3\pi}{2}$"]

ax.set_xticks(xticker)
ax.set_xticklabels(xlabels, size=17)

ax.text(np.pi, 1.1, "y=sin(x)")

ax.set_ylim(-1.5, 1.5)
yticker = np.arange(-1, 2, 1)
ax.set_yticks(yticker)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab

plt.figure(figsize=(15,3))

plt.subplot(121)
mean = 175
variance = 3000
sigma = np.sqrt(variance)
x = np.linspace(0,365,365)
ts1 = pd.Series(mlab.normpdf(x,mean,sigma), index=pd.date_range('1/1/2014', periods=365))

mean = 195
variance = 3500
sigma = np.sqrt(variance)
x = np.linspace(0,365,365)
ts2 = pd.Series(mlab.normpdf(x,mean,sigma), index=pd.date_range('1/1/2014', periods=365))
ts1.plot(ylim=[ts1.max()*-1-(0.1*ts1.max()),ts1.max()+(0.1*ts1.max())], linewidth="3", alpha=0.3, grid=False)
ts2.plot(linewidth="3", alpha=0.3, grid=False)

plt.subplot(122)
mean = 175
variance = 2500
sigma = np.sqrt(variance)
x = np.linspace(0,365,365)
ts1 = pd.Series(mlab.normpdf(x,mean,sigma)*-1, index=pd.date_range('1/1/2014', periods=365))

mean = 220
variance = 3000
sigma = np.sqrt(variance)
x = np.linspace(0,365,365)
ts2 = pd.Series(mlab.normpdf(x,mean,sigma)*-1, index=pd.date_range('1/1/2014', periods=365))
ts1.plot(ylim=[ts1.min()+(0.1*ts1.min()),ts1.min()*-1-(0.1*ts1.min())],linewidth="3", alpha=0.3)
ts2.plot(linewidth="3", alpha=0.3)
```
