```python
files = dl.FilesFolder(r'D:\Data\GPCP_anomaly')
ds = gdal.Open(files[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

max_ds = gdal.Open(files[0]).ReadAsArray()
max_ds = np.ma.masked_values(max_ds, nan)
min_ds = gdal.Open(files[0]).ReadAsArray()
min_ds = np.ma.masked_values(min_ds, nan)
for i in files:
    
    ds_comp = gdal.Open(i).ReadAsArray()
    ds_comp = np.ma.masked_values(ds_comp, nan)
    
    max_ds = np.maximum(max_ds, ds_comp)
    min_ds = np.minimum(min_ds, ds_comp)
    print (max_ds.max(), min_ds.min())
    
    
dl.saveRaster(r'D:\Data\GPCP_maxmin//GPCP_max.tif', max_ds.data, ds, datatype=6, nan=nan)
dl.saveRaster(r'D:\Data\GPCP_maxmin//GPCP_min.tif', min_ds.data, ds, datatype=6, nan=nan)
```

```python
pathmax = r'D:\Data\GPCP_maxmin//GPCP_max.tif'
pathmin = r'D:\Data\GPCP_maxmin//GPCP_min.tif'
pathano = r'D:\Data\GPCP_anomaly'

fileano = dl.FilesFolder(pathano, '.tif')

# open basefile
ds = gdal.Open(fileano[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

# open max and min files, mask where NaN
maxin = gdal.Open(pathmax).ReadAsArray()
maxin = np.ma.masked_values(maxin, nan)
minin = gdal.Open(pathmin).ReadAsArray()
minin = np.ma.masked_values(minin, nan)

for i in fileano:
    # open anomaly and mask where NaN
    ano = gdal.Open(i).ReadAsArray()
    ano = np.ma.masked_values(ano, nan)
    
    # normalized anomaly index
    normano = ano / (maxin - minin)
    
    print ('max', normano.max(), 'min', normano.min(), 'std', normano.std())
    path = r'D:\Data\GPCP_anomalyNorm'+i[-17::]
    print (path)
    
    dl.saveRaster(path, normano.data, ds, datatype=6, nan=nan)
```

```python
path = r'D:\Data\GPCP_doy8DayMean'
file8daymean = dl.FilesFolder(path, '.tif')

# open basefile
ds = gdal.Open(file8daymean[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()
if nan == None:
    nan = -99999.0
arr = band.ReadAsArray()
    
for start in range(1,366,8):
    
    # set start and end doy    
    end = start + 30
    print ('\nstart', str(start).zfill(3), 'end', str(end).zfill(3))
    
    for year in range(2000,2014):
        print (year)
        
        # select all files for each year        
        annual = []        
        for fil in file8daymean:
            if int(fil[-11:-7]) == year:
                annual.append(fil)
        
        # create base_array        
        base_array = np.full_like(arr, nan)        
        base_array = base_array[None]
        
        # for each year within start and end doy hstack all available rasters        
        for filannual in annual:
            if start <= int(filannual[-7:-4]) <= end:
                
                # first try if raster is available                
                try:
                    print (filannual)
                    array = gdal.Open(filannual).ReadAsArray()
                    array = array[None]
                    base_array = np.vstack((base_array, array))
         
                except IndexError:
                    pass
                continue
                
            # mask nan values and calculate sum/accumulation over axis 0        
        base_array = np.ma.masked_values(base_array, nan)
        base_array = base_array.mean(axis=0)
        base_array = base_array.sum(axis=0)
        
        print ('max',base_array.max(),'min', base_array.min(), 'std', base_array.std(), 'mean', base_array.mean() )

        # set outpath
        folder = r'D:\Data\GPCP_30acc//'
        filename = 'GPCP_'+str(year)+str(start).zfill(3)+'.tif'
        path = folder+filename
        print (path)

        #print(base_array.data)
        #dikkelul.saveRaster(path, base_array.data, ds, 6, nan=nan)                

        base_array = None
        array = None
    year = None
```

```python
files = dl.FilesFolder(r'D:\Data\GPCP_30acc')
ds = gdal.Open(files[0])
band = ds.GetRasterBand(1)
nan = band.GetNoDataValue()

# Open two basefiles, one for max one for min, mask arrays where NaN
max_ds = gdal.Open(files[0]).ReadAsArray()
max_ds = np.ma.masked_values(max_ds, nan)
min_ds = gdal.Open(files[0]).ReadAsArray()
min_ds = np.ma.masked_values(min_ds, nan)

for i in files:
    
    # Open file, mask where NaN. 
    ds_comp = gdal.Open(i).ReadAsArray()
    ds_comp = np.ma.masked_values(ds_comp, nan)
    
    # Compare two arrays and returns a new array containing the element-wise maxima
    max_ds = np.maximum(max_ds, ds_comp)
    min_ds = np.minimum(min_ds, ds_comp)
    
    print (max_ds.max(), min_ds.min())    
    
dl.saveRaster(r'D:\Data\GPCP_30acc_maxmin//GPCP_30acc_max.tif', max_ds.data, ds, datatype=6, nan=nan)
dl.saveRaster(r'D:\Data\GPCP_30acc_maxmin//GPCP_30acc_min.tif', min_ds.data, ds, datatype=6, nan=nan)
```

### Annual Mean of 30-day moving window

```python
path = r'D:\Data\GPCP_30acc'
files = dl.FilesFolder(path, '.tif')

ds = gdal.Open(files[0])
band = ds.GetRasterBand(1)
barray = band.ReadAsArray()
nan = band.GetNoDataValue()

for i in range(1,366,8):
    #print (i)
    base = np.full_like(barray, nan)
    base = base[None] # add another dimenstion for vstack
    for fil in files:    
        
        if int(fil[-7:-4]) == i:
            print (fil)
            array = gdal.Open(fil).ReadAsArray().astype(float)

            array = array[None] # add another dimension for vstack
            base = np.vstack((base,array)) # vstack all rasters with same doy
            
    # mask where value is nan
    base = np.ma.masked_equal(base, nan)
    base = np.mean(base, axis=0)
    
    outfolder = r'D:\Data\GPCP_30acc_annualMean//'
    outpath = outfolder+'GPCP_'+str(i).zfill(3)+'.tif'
    print (outpath)
    
    dl.saveRaster(outpath, base, ds, 6, nan=nan)
```
