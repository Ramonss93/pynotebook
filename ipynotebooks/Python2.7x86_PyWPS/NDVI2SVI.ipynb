Calculate the simple Standardized Vegetation Index using a z-score as follow: $$$$ $$z_{ijk} = \frac{NDVI_{ijk} -  \overline{NDVI}_{ij}}{\sigma_{ij}} $$ $$$$
Where $z_{ijk}$ is the z-value for pixel $i$ during day $j$ for year $k$. $NDVI_{ijk}$ is the daily $NDVI$ value for pixel $i$ during day $j$ for year $k$, $\overline{NDVI}_{ij}$ is the mean $NDVI$ for pixel $i$ during day $j$ over $n$ years, and sigma_{ij} is the standard deviation of pixel $i$ during day $j$ over $n$ years

```python
svi = ndvi.groupby([ndvi.index.month, ndvi.index.day]).apply(lambda g: (g - g.mean())/g.std())
svi.columns = ['SVI']
svi.head()
```
