```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy import fftpack as fft
```

#### Creation of a test signal

```python
n = 100
t = (2.0 * pi / n) * arange(n)
dt = t[1] - t[0]
s = 0.5 * t \
    + sin(4.0 * t) \
    + (0.5 * cos(10.5 * t)) \
    + (0.5 * np.random.randn(n)) #int(0,100)
```

#### Display the frequencies

```python
freq_lowest = 1/(n*dt)
freq_nyquist = 1.0 / (2.0 * dt)
freq = arange(n/2.0 + 1.0) * freq_lowest
print freq_lowest, freq_nyquist
```

#### Apply a Line of regression through original function and  subtract it from the original function - DETRENDING

```python
c = stats.linregress(t, s)
s_fit = c[0] + c[1] * t
s_dtr = s - s_fit
```

#### Plotting original, detrended and regression

```python
plt.plot(t,s, 'g-') # Original function
plt.plot(t,s_dtr, 'r--') # Detrended function of original
plt.plot(t,s_fit, 'b:') # Regression of original
plt.ylim(-2,4)
plt.xlim(0,2*pi)
plt.show()
```
