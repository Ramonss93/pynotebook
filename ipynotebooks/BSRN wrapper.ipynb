```python
import pandas as pd
import os
from datetime import datetime
import numpy as np
```

```python
def funcFiles(infolder):
    files_ = [os.path.join(root, name)
              for root, dirs, files in os.walk(infolder)
              for name in files
              if name.endswith('.tab')]
    return files_
```
