```python
da = []
va = []
path = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\Jiujang\NDVI\RAW2'
files = FilesFolder(path, format_end='tif')
for i in files:
    ds = gdal.Open(i).ReadAsArray()
    nameRas = i[-11:-4]
    date = datetime.datetime(int(nameRas[0:4]), 1, 1) + datetime.timedelta(int(nameRas[4:9]) - 1) # replaced "int(nameRas[0:4])" with "2010" first year then add doy
    da.append(date)
    
    ar = np.ma.masked_equal(ds, ds.min()).mean()
    va.append(ar)
ts = pd.Series(data = va, index = da)
ts.plot()
```
